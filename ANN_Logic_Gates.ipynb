### Basic ANN Program

```python
import torch
```

```python
import torch.nn as nn
```

```python
class SimpleNN(nn.Module):
    def __init__(self,input_size,hidden_size,output_size) :
        super().__init__()
        self.l1=nn.Linear(input_size,hidden_size)
        self.relu=nn.ReLU()
        self.l2=nn.Linear(hidden_size,output_size)

    def forward(self,x):
        x=self.l1(x)
        x= self.relu(x)
        x=self.l2(x)
        return x
```

```python
model=SimpleNN(input_size=4,hidden_size=5,output_size=3)
```

```python
x=torch.tensor([[1.0,2.2,4.4,3.3]])
output=model(x)
```

### AND GATE Model Building

```python
import torch
import torch.nn as nn
```

```python
class ANDAGATE(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_fc=nn.Linear(2,1)
    def forward(self,x):
        x=self.layer_fc(x)
        x=torch.sigmoid(x)
        return x
```

```python
model = ANDAGATE()
```

```python
x = torch.tensor([[0.,0.],[0.,1.],[1.0,0.0],[1.0,1.0]])
y = torch.tensor([[0.],[0.],[0.],[1.]])
```

```python
loss_counter = nn.BCELoss()
```

```python
optimizer = torch.optim.SGD(model.parameters(),lr=0.1)
```

```python
for epoch in range(400):

    optimizer.zero_grad() #reset old gradient

    output=model(x)  #forward pass

    loss = loss_counter(output,y) # compute loss

    loss.backward()  # backward Propagation

    optimizer.step() # update Weights

    if epoch % 200 == 0:
        print(epoch, loss.item())
```

### XOR GATE WITH ANN 

```python
import torch
import torch.nn as nn
```

```python
c = torch.tensor([[0.,0.],[0.,1.],[1.,0.],[1.,1.]])

d = torch.tensor([[0.],[1.],[1.],[0.]])
```

```python
print(f"Shape Of C : {c.shape}")
print(f"Shape of D :{d.shape}")
```

```python
class XORGATE(nn.Module):
    def __init__(self,input_size,output_size,hiddenlayer_nueron):
        super().__init__()
        self.hidden_layer = nn.Linear(input_size,hiddenlayer_nueron)
        self.output_layer = nn.Linear(hiddenlayer_nueron, output_size)

    def forward(self,c):
        activation_fn = torch.sigmoid(self.hidden_layer(c))

        c = torch.sigmoid(self.output_layer(activation_fn))
        return c

XORModel = XORGATE(input_size=2,output_size=1,hiddenlayer_nueron=2)
```

```python
loss_count3 = nn.BCELoss()
```

```python
optimizer3 = torch.optim.SGD(XORModel.parameters(),lr=0.5)
```

```python
for epoch in range(22000):
    optimizer3.zero_grad()

    outputs4 = XORModel(c)

    loss = loss_count3(outputs4,d)

    loss.backward()

    optimizer3.step()

    if epoch % 1000 == 0 :
        print(epoch,loss.item())
```

```python
print(XORModel(c))
```

### NOR GATE WITH ANN

```python
import torch
import torch.nn
```

```python
input_x = torch.tensor([[0.,0.],[0.,1.],[1.,0.],[1.,1.]])
output_y = torch.tensor([[1.],[0.],[0.],[0.]])
```

```python
class NORGATE(nn.Module):
    def __init__(self):
        super().__init__()
        self.input_layer  = nn.Linear(2,2)
        self.output_layer = nn.Linear(2,1)

    def forward(self,input_x):
        activation_fn = torch.sigmoid(self.input_layer(input_x))
        input_x = torch.sigmoid(self.output_layer(activation_fn))
        return input_x
```

```python
NORModel = NORGATE()
```

```python
loss_count4 = nn.BCELoss()
```

```python
optimizer4 = torch.optim.SGD(NORModel.parameters(),lr=0.5)
```

```python
for epoch in range(20000):
   optimizer4.zero_grad()
   outputs5 = NORModel(input_x)
   loss = loss_count4(outputs5, output_y)
   loss.backward()
   optimizer4.step()

   if epoch % 100 == 0 :
        print(epoch,loss.item())
```

```python
print(NORModel(input_x))
```
