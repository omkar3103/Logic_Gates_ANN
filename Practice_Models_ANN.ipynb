### AND GATE PRACTICE IN ANN

```python
import torch
import torch.nn as nn
```

```python
x= torch.tensor([[0.,0.],[0.,1.],[1.,0.],[1.,1.]])
y= torch.tensor([[0.],[0.],[0.],[1.]])
```

```python
print(x.shape)
print(y.shape)
```

```python
class ANDGate(nn.Module):
    def __init__(self):
        super().__init__()
        self.input_layer = nn.Linear(2,4)
        self.relu = nn.ReLU()
        self.output_layer = nn.Linear(4,1)
        self.sigmod = nn.Sigmoid()
    
    def forward(self,x):
        x = self.input_layer(x);
        x = self.relu(x)
        x = self.output_layer(x)
        x = self.sigmod(x)
        return(x)
```

```python
AndGateModel = ANDGate()
```

```python
loss_count = nn.BCELoss()
```

```python
optimizer = torch.optim.SGD(AndGateModel.parameters(),lr=0.5)
```

```python
loss_values = []
for epoch in range(4000):

    optimizer.zero_grad()

    outputs = AndGateModel(x)

    loss = loss_count(outputs,y)

    loss.backward()

    optimizer.step()

    loss_values.append(loss.item())

    if epoch % 20 == 0:
        print(epoch,loss.item())
```

```python
print(AndGateModel(x))
```

### Plottimg on Graph

```python
import matplotlib.pyplot as plt

plt.plot(loss_values)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("AND Gate Training Loss Curve")
plt.show()
```

### OR GATE WITH ANN

```python
import torch
import torch.nn as nn
```

```python
inputs_x = torch.tensor([[0.,0.],[0.,1.],[1.,0.],[1.,1.]])
output_y = torch.tensor([[0.],[1.],[1.],[1.]])
```

```python
class ORGATE(nn.Module):
    def __init__(self):
        super().__init__()
        self.input_layer= nn.Linear(2,4)
        self.relu2=nn.ReLU()
        self.hidden_layer= nn.Linear(4,1)
        self.sigmod2=nn.Sigmoid()

    def forward(self,inputs_x):
        inputs_x = self.input_layer(inputs_x)
        inputs_x = self.relu2(inputs_x)
        inputs_x = self.hidden_layer(inputs_x)
        inputs_x = self.sigmod2(inputs_x)
        
        return (inputs_x)
        
```

```python
ORGATEModel = ORGATE()
```

```python
loss_count2 = nn.BCELoss()
```

```python
optimizer2 = torch.optim.SGD(ORGATEModel.parameters(),lr=0.2)
```

```python
loss_values=[]
for epoch in range (4000):
    optimizer2.zero_grad()

    outputs2 = ORGATEModel(inputs_x)

    loss = loss_count2(outputs2,output_y)

    loss.backward()

    optimizer2.step()


    loss_values.append(loss.item())
    
    if epoch % 200 == 0:
        print(epoch,loss.item())
```

```python
print(ORGATEModel(inputs_x))
```

```python
import matplotlib.pyplot as plt
plt.plot(loss_values)
plt.xlabel('epochs')
plt.ylabel('loss')
```

### NAND GATE WITH ANN 

```python
import torch
import torch.nn as nn
```

```python
input_a = torch.tensor([[0.,0.],[0.,1.],[1.,0.],[1.,1.]])
output_b = torch.tensor([[1.],[1.],[1.],[0.]])
```

```python
class NANDGATE(nn.Module):
    def __init__(self):
        super().__init__()
        self.input_layer = nn.Linear(2,4)
        self.relu2 = nn.ReLU()
        self.hidden_layer = nn.Linear(4,4)
        self.output_layer = nn.Linear(4,1)
        self.sigmod2 = nn.Sigmoid()
        
    def forward(self,input_a):
        input_a = self.input_layer(input_a)
        input_a = self.relu2(input_a)
        input_a = self.hidden_layer(input_a)
        input_a= self.relu2(input_a)
        input_a = self.output_layer(input_a)
        input_a = self.sigmod2(input_a)

        return (input_a)
```

```python
NANDGATEModel = NANDGATE()
```

```python
loss_count3 = nn.BCELoss()
```

```python
optimizer3 = torch.optim.SGD(NANDGATEModel.parameters(),lr=0.4)
```

```python
loss_values=[]
for epoch in range (1800):
    
    optimizer3.zero_grad()
    
    outputs3 = NANDGATEModel(input_a) 
    
    loss = loss_count3(outputs3,output_b)
    
    loss.backward()
    
    optimizer3.step()
    
    loss_values.append(loss.item())

    if epoch % 200 == 0 :
        print(epoch,loss.item())
```

```python
print(NANDGATEModel(input_a))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(loss_values)
plt.xlabel('epochs')
plt.ylabel('loss')
```

### NOR GATE WITH ANN

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from torchview import draw_graph
```

```python
input_nor_x = torch.tensor([[0.,0.],[0.,1.],[1.,0.],[1.,1.]])

output_nor_y = torch.tensor([[1.],[0.],[0.],[0.]]) 
```

```python
class XORGATE(nn.Module):
    def __init__(self):
        super().__init__()
        self.input_layer = nn.Linear(2,24)
        self.relu = nn.ReLU()
        self.hidden_layer = nn.Linear(24,12)
        self.hidden_layer2 = nn.Linear(12,6)
        self.output_layer_nor = nn.Linear(6,1)
        self.sigmod4 = nn.Sigmoid()

    def forward(self,input_nor_x):
        input_nor_x = self.input_layer(input_nor_x)
        input_nor_x = self.relu(input_nor_x)
        input_nor_x = self.hidden_layer(input_nor_x)
        input_nor_x = self.relu(input_nor_x)
        input_nor_x = self.hidden_layer2(input_nor_x)
        input_nor_x = self.relu(input_nor_x)
        input_nor_x = self.output_layer_nor(input_nor_x)
        input_nor_x = self.sigmod4(input_nor_x)

        return(input_nor_x)
```

```python
XORMOdel = XORGATE()
```

```python
loss_count_nor = nn.BCELoss()
```

```python
optimizer_nor = torch.optim.Adam(XORMOdel.parameters(),lr=0.3)
```

```python
loss_values = []

epochs = 400
for epoch in range (epochs):
    optimizer_nor.zero_grad()

    outputs_nor = XORMOdel(input_nor_x)

    loss = loss_count_nor(outputs_nor, output_nor_y)

    loss.backward()

    optimizer_nor.step()

    loss_values.append(loss.item())

    if epoch % 200 == 0 :
        print(epoch,loss.item())
```
